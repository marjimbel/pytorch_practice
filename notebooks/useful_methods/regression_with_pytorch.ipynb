```python
import torch
 
from torch.utils import data 
from torch import nn
```

```python
# generate synthetic data

def synthetic_data(c, num_examples):
    X = torch.randn(num_examples, 2)
    y = (X[:,0]*c + 2) + torch.sin(X[:,1]) + (3*X[:,0]+.001*X[:,1])
    return X, y.reshape(-1,1)

features, labels  = synthetic_data(3, 1000)
print(features.shape)
print(labels.shape)
```

```python
def load_array(data_arrays, batch_size, is_train=True):
    """Construct a PyTorch data iterator."""
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle=is_train)
 
batch_size = 10
data_iter = load_array((features, labels), batch_size)
x, y = next(iter(data_iter))
print(x.shape)
print(y.shape)
```

```python
class LNN(nn.Module):
    def __init__(self):
        super(LNN, self).__init__()
        self.fc1 = nn.Linear(2,16)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(16,1)
    
    def forward(self,x):
        x = self.fc1(x)
        x = self.relu(x)
        out = self.fc2(x)
        return out

model = LNN()        
print(model)
model.forward(x)
```

```python
epochs = 10

model = LNN()        
criteria = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=.001)

for e in range(0, epochs):
    features, labels  = synthetic_data(3, 1000)
    data_iter = load_array((features, labels), batch_size)
    for batch, (x,y) in enumerate(data_iter):
        
        optimizer.zero_grad()
        preds = model.forward(x)
        loss = criteria(preds,y)
        loss.backward()
        optimizer.step()
        
        if batch%10==0:
            print(f"loss in epoch:{e} and batch:{batch} is {loss}")        
            
    plot_test_set()  

```

```python
import matplotlib.pyplot as plt

def plot_test_set():
    test_features, test_labels = synthetic_data(3, 100)

    preds = model(test_features)
    preds = preds.squeeze().tolist()

    plt.plot(preds, color="red", label="preds")
    plt.plot(test_labels, color="blue", label="true")
    plt.legend(loc="best")
    plt.grid()
    plt.show()
    
plot_test_set()  
    
```

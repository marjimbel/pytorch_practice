```python
#indexing
arr = np.arange(0,10)
print(arr[8:10])
print(arr.item(8))
```

```python
# indexing a 2D array
array_2d = np.arange(20).reshape(4,5)
assert(array_2d[3][4]==19)
assert(array_2d[3,4]==19)

i, j = np.where(array_2d == 19)
assert(array_2d[i,j]==19)

i, j = np.where(array_2d == np.amax(array_2d))
assert(array_2d[i,j]==19)
```

```python
# slicing
array_2d = np.arange(5,50,5).reshape(3,3)
# get second 2x2 filter with stride = 1
print(array_2d[:2,1:])
array_2d[:2][:,1:]
```

```python
# mask
arr = np.random.rand(10)
arr[arr>.5]
```

```python
# central limit theorem
ls1 = []
ls2 = []
for i in range(0,100):
    dims = 1000
    arr = np.random.rand(dims,dims)
    l = arr[arr>.5].shape[0]/dims**2
    ls1.append(l)

for i in range(0,100):
    dims = 100
    arr = np.random.rand(dims,dims)
    l = arr[arr>.5].shape[0]/dims**2
    ls2.append(l)

plt.hist(ls1, alpha=.6)
plt.hist(ls2, alpha=.6)
plt.legend(loc='best')
plt.grid()
```
